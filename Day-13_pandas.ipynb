# **Joins**

```python
import pandas as pd
import numpy as np
```

```python
nov = pd.read_csv('../lecture13/reg-month1.csv')
dec = pd.read_csv('../lecture13/reg-month2.csv')
```

```python
courses = pd.read_csv('../lecture13/courses.csv')
students = pd.read_csv('../lecture13/students.csv')
```

```python
reg = pd.concat([nov, dec], ignore_index=True)
```

```python
reg.head()
```

## Merge

```python
students.head()
```

```python
students
```

```python
reg
```

```python
# inner join
students.merge(reg, how='inner', on='student_id')
```

```python
courses
```

```python
# left join
reg.merge(courses, how='left', on='course_id')
```

```python
# right join
df = pd.DataFrame({
    'student_id':[26, 27, 28],
    'name':['Hasnain','Muhammad','Daniel'],
    'partner':[28, 26, 17]
})

students = pd.concat([students, df],ignore_index=True)
```

```python
students
```

```python
reg
```

```python
reg
```

```python
students.merge(reg , how='right', on='student_id')
```

```python
# outer join
students.merge(reg, how='outer', on='student_id').tail(10)
```

```python
reg.sample()
```

```python
courses.sample()
```

```python
students.sample()
```

```python
# find total revenue generated
revenue = courses.merge(reg, how='inner', on='course_id')['price'].sum()
revenue
```

```python
# find month by month revenue
months = pd.concat([nov, dec], keys=['November', 'December']).reset_index()
```

```python
courses.merge(months, how='inner', on='course_id').groupby('level_0')['price'].sum()
```

```python
# Print the registration table
# cols -> name -> course -> price
reg.merge(students, on='student_id').merge(courses, on='course_id')[['name','course_name','price']]
```

```python
# Plot bar chart for revenue/course
reg.merge(courses, on='course_id').groupby('course_name')['price'].sum().sort_values(ascending=False).plot(kind='bar')
```
